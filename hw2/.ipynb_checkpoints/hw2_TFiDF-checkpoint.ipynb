```python
import jieba
import jieba.analyse
import os
import sys
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
os.chdir("C:\\Users\\steven\\Desktop\\2018Summer\\python\\summer_python\\hw2")
jieba.set_dictionary('dict.txt.big.txt')
```

```python
Food = ''
Food_list = []

os.chdir("C:\\Users\\steven\\Desktop\\2018Summer\\python\\summer_python\\hw2\\Food")

for i in range(1,101):
    content = open('F_{0}.txt'.format(i),'r', encoding = 'utf8').read()
#     content = content.replace("\n","")
#     content = content.replace(" ","")
#     content = content.replace(",","")
#     content = content.replace("(","")
#     content = content.replace(")","")
#     content = content.replace("-","")
#     content = content.replace("[","")
#     content = content.replace("]","")
#     content = content.replace(".","")
    
    words = jieba.cut(content,cut_all=False)
    
    temp_word = ""
    for word in words:
        temp_word = temp_word + word
        temp_word = temp_word + " "
    
    
#     temp_word = temp_word.strip() 
    Food_list.append(temp_word)
    
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(Food_list)
# print (tfidf.shape)

wordOfFood = vectorizer.get_feature_names()

# # 輸出
# for i in range(len(Food_list)):
#     print ('----Document %d----' % (i))
#     for j in range(len(wordOfFood)):
#         if tfidf[i,j] > 0.8:
#             print (wordOfFood[j], tfidf[i,j])
```

```python
tfidf_np_Food = tfidf.toarray()
Max_60words_Food = []
```

```python
for i in range(0,60):
    maxx = np.where(tfidf_np_Food==np.max(tfidf_np_Food))
#     print(wordOfFood[maxx[1][0]] , tfidf[maxx[0][0],maxx[1][0]])
    Max_60words_Food.append(maxx[1][0])
    for j in range(0,100):
        tfidf_np_Food[j,maxx[1][0]] = 0
    i = i+1
```

```python
Food_60 = []
for i in Max_60words_Food :
    Food_60.append(wordOfFood[i])
```

```python
os.chdir("C://Users//steven//Desktop//2018Summer//python//summer_python//hw2//Mobile")

Mobile = ''
Mobile_list = []


for i in range(1,101):
    content = open('M_{0}.txt'.format(i),'r', encoding = 'utf8').read()
#     content = content.replace("\n","")
#     content = content.replace(" ","")
#     content = content.replace(",","")
#     content = content.replace("(","")
#     content = content.replace(")","")
#     content = content.replace("-","")
#     content = content.replace("[","")
#     content = content.replace("]","")
#     content = content.replace(".","")
    
    words = jieba.cut(content,cut_all=False)
    
    temp_word = ""
    for word in words:
        temp_word = temp_word + word
        temp_word = temp_word + " "
    
    
#     temp_word = temp_word.strip() 
    Mobile_list.append(temp_word)
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(Mobile_list)
# print (tfidf.shape)

wordOfMobile = vectorizer.get_feature_names()

# 輸出
# for i in range(len(Food_list)):
#     print ('----Document %d----' % (i))
#     for j in range(len(wordOfFood)):
#         if tfidf[i,j] > 0.8:
#             print (wordOfFood[j], tfidf[i,j])
```

```python
tfidf_np_Mobile = tfidf.toarray()
Max_60words_Mobile = []
```

```python
for i in range(0,60):
    maxx = np.where(tfidf_np_Mobile==np.max(tfidf_np_Mobile))
#     print(wordOfMobile[maxx[1][0]] , tfidf[maxx[0][0],maxx[1][0]])
    Max_60words_Mobile.append(maxx[1][0])
    for j in range(0,100):
        tfidf_np_Mobile[j,maxx[1][0]] = 0
    i = i+1
```

```python
Mobile_60 = []
for i in Max_60words_Mobile :
    Mobile_60.append(wordOfMobile[i])
```

```python
os.chdir("C://Users//steven//Desktop//2018Summer//python//summer_python//hw2/Integrated/")

#initial
Check_att = []
Check_att = Food_60 + Mobile_60

temp_att = []
Check_Boo = []
temp_att.append(" ")

for w in Check_att:
    temp_att.append(w)
    
Check_Boo.append(temp_att)


#Food
for i in range(1,101):
    content = open('F_{0}.txt'.format(i),'r', encoding = 'utf8').read()

    words = jieba.cut(content,cut_all=False)
    
    temp_list = []
    temp_check = []
    temp_check.append("F_{0}".format(i))
    for word in words:
        temp_list.append(word)
    
    for j in range(0,120):
        if Check_att[j] in temp_list:
            temp_check.append("1")
        elif Check_att[j] not in temp_list:
            temp_check.append("0")
    
    Check_Boo.append(temp_check)
    
    
    
#Mobile
for i in range(1,101):
    content = open('M_{0}.txt'.format(i),'r', encoding = 'utf8').read()

    words = jieba.cut(content,cut_all=False)
    
    temp_list = []
    temp_check = []
    temp_check.append("M_{0}".format(i))
    for word in words:
        temp_list.append(word)
    
    for j in range(0,120):
        if Check_att[j] in temp_list:
            temp_check.append("1")
        elif Check_att[j] not in temp_list:
            temp_check.append("0")
    
    Check_Boo.append(temp_check)
```
